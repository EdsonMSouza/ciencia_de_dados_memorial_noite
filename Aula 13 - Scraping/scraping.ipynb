```
import requests as req # biblioteca para realizar requisições WEB
from bs4 import BeautifulSoup # Realiza tratamentos HTML
```

```
url = "http://www.uol.com.br"
pagina = req.get(url) # pega o conteúdo de uma página e coloca na variável página
```

```
# erificar se houve uma resposta do servidor
pagina.status_code
```

```
pagina.text # mostra o conteúdo da pagina
```

```
bs = BeautifulSoup(pagina.text, 'html.parser')
```

```
# mostrar o conteúdo de uma forma mais legível
print(bs.prettify())
```

```
# Mostrar somente o conteúdo de TAGS 'p'
bs.find_all('div')
```

```
array = bs.find_all('div')

for x in array:
  print(x.get_text())
```

```
# mostrando conteúdos dentro de listas 'ul>li'
x = bs.findAll('ul')
for ul in x:
  for li in ul.findAll('li'):
    print(li.get_text())
```

```
# procurando mais de um elemento de uma só vez
bs.find_all(['a', 'div'])
```

```
# Resolvendo a parada: colocar os valores em arrays - a, div e p
# strip() remove os espaços iniciais de uma cadeia de texto
a = []
for tmp in bs.find_all('a'):
  # testar se a linha (conteúdo) é vazio
  if len(tmp.get_text().strip()) != 0:
    a.append(tmp.get_text().strip())

div = []
for tmp in bs.find_all('div'):
  # testar se a linha (conteúdo) é vazio
  if len(tmp.get_text().strip()) != 0:
    div.append(tmp.get_text().strip())

p = []
for tmp in bs.find_all('p'):
  # testar se a linha (conteúdo) é vazio
  if len(tmp.get_text().strip()) != 0:
    p.append(tmp.get_text().strip())
```
